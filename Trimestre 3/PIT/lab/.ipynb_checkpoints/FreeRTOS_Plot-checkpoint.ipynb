```python
import serial, time
import numpy as np
import matplotlib.pyplot as plt
```

```python

arduino = serial.Serial('/dev/ttyUSB0', 115200)
time.sleep(2)
```

```python
'''M = input("Ciclo Principal: ")

periodo_tarea_1 = input("Periodo Tarea 1: ")
periodo_tarea_2 = input("Periodo Tarea 2: ")
periodo_tarea_3 = input("Periodo Tarea 3: ")

tast_to_send = str(periodo_tarea_1) + " " + str(periodo_tarea_2) + " " + str(periodo_tarea_3)
aa = tast_to_send.encode()
arduino.write(aa)'''

M = 100
tarea_1 = []
tarea_2 = []
tarea_3 = []

dd = []
ee = []
ff = []

conta_tarea_1 = 0
conta_tarea_1_a = 0

conta_tarea_2 = 0
conta_tarea_2_a = 0

conta_tarea_3 = 0
conta_tarea_3_a = 0

conta_tarea_4 = 0
conta_tarea_4_a = 0
```

```python


for i in range(int(M)):    

    rawString = arduino.readline()
    
    get_data = rawString.decode("utf-8")     
    
    if "tarea_1" in get_data:
        dd.append(get_data)
    elif "tarea_2" in get_data:
        ee.append(get_data)        
    elif "tarea_3" in get_data:
        ff.append(get_data)  
        
    # Tarea 1   
    if "tarea_1" in get_data and "tarea_2" != get_data and "tarea_3" != get_data:
        conta_tarea_1 = conta_tarea_1 + 1  
    else:
        conta_tarea_1_a = conta_tarea_1_a + 1        
    j = 0
    for i in range(conta_tarea_1):
        tarea_1.append(3)
        j = i
    conta_tarea_1 = 0
    
    for j in range(conta_tarea_1_a):
        tarea_1.append(2)        
    conta_tarea_1_a = 0
            
        
    # Tarea 2       
    if "tarea_2" in get_data and "tarea_1" != get_data and "tarea_3" != get_data:
        conta_tarea_2 = conta_tarea_2 + 1        
    else:
        conta_tarea_2_a = conta_tarea_2_a + 1        
        
    for i in range(conta_tarea_2):
        tarea_2.append(2)        
    conta_tarea_2 = 0   
    
    for i in range(conta_tarea_2_a):
        tarea_2.append(1)        
    conta_tarea_2_a = 0

        
    # Tarea 3
    if "tarea_3" in get_data and "tarea_2" != get_data and "tarea_1" != get_data:
        conta_tarea_3 = conta_tarea_3 + 1        
    else:
        conta_tarea_3_a = conta_tarea_3_a + 1        
        
    for i in range(conta_tarea_3):
        tarea_3.append(1)
    conta_tarea_3 = 0  
    
    for i in range(conta_tarea_3_a):
        tarea_3.append(0)        
    conta_tarea_3_a = 0
       

        
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.rcParams['figure.figsize'] = [50, 20]
plt.rcParams.update({'font.size': 10})

tarea_1_patch = mpatches.Patch(color='blue', label='Tarea 1')
plt.plot(tarea_1, c='b') 

tarea_2_patch = mpatches.Patch(color='red', label='Tarea 2')
plt.plot(tarea_2,c='r') 

tarea_3_patch = mpatches.Patch(color='green', label='Tarea 3')
plt.plot(tarea_3,c='g') 

plt.legend(handles=[tarea_1_patch, tarea_2_patch, tarea_3_patch])
plt.xticks(range(0, int(M)))

plt.grid()
plt.show()
```

```python


tareas = tarea_1 + tarea_2 + tarea_3

aa = []
bb = []
cc = []

tareas = dd + ee + ff
for i in tareas:
    if "tarea_1" in i:
        aa.append(3)
    else:
        aa.append(2)

    if "tarea_2" in i:
        bb.append(2)
    else:
        bb.append(1)        

    if "tarea_3" in i:
        cc.append(1)
    else:
        cc.append(0)    
```

```python

plt.rcParams['figure.figsize'] = [50, 20]
plt.rcParams.update({'font.size': 10})

tarea_1_patch = mpatches.Patch(color='blue', label='Tarea 1')
plt.plot(aa, c='b') 

tarea_2_patch = mpatches.Patch(color='red', label='Tarea 2')
plt.plot(bb,c='r') 

tarea_3_patch = mpatches.Patch(color='green', label='Tarea 3')
plt.plot(cc,c='g') 

plt.legend(handles=[tarea_1_patch, tarea_2_patch, tarea_3_patch])
plt.xticks(range(0, int(M)))

plt.grid()
plt.show()
```
