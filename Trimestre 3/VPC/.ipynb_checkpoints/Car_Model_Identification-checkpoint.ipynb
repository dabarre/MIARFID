```python
def configure_callbacks(model_id, kpi_to_monitor='val_accuracy'):
    # Without log/ or models/ subfolder as not possible to access unexisting folders
    # If possible to train with jupyter revise
    path = "models_cars"
    name = "cmi"
    
    log_filename = '%s/%s-%s.log' % (path,name, model_id)
    csv_logger = CSVLogger(log_filename)
    
    chk_1_model_filename = '%s/%s-%s-{epoch:04d}-{%s:.6f}.h5' % (path, name, model_id, kpi_to_monitor)
    chk_2_model_filename = '%s/%s-%s.h5' % (path, name, model_id)
    
    # Save best model fully not only weights after each epoch (period=1) 
    # with best accuracy value (mode=max, save_best_only=True)
    checkpoint1 = ModelCheckpoint(
        chk_1_model_filename,
        monitor=kpi_to_monitor,
        save_best_only=True,
        save_weights_only=False, 
        verbose=1, mode='max', period=1
    )

    checkpoint2 = ModelCheckpoint(
        chk_2_model_filename, 
        monitor=kpi_to_monitor,
        save_best_only=False,
        save_weights_only=False, 
        verbose=1, mode='auto', period=1
    )

    callbacks = [csv_logger, checkpoint1, checkpoint2]
    return callbacks
```

```python
# Data Augmentation with an ImageGenerator + more variations
datagen = ImageDataGenerator(width_shift_range=0.2,
                             height_shift_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=False,
                             zoom_range=[1.0,1.2],
                             rotation_range=20.0,
                             shear_range=0.1)

def multiple_data_generator(generator, X, Y, bs):
    genX = generator.flow(X, Y,batch_size=bs)
    while True:
      [Xi,Yi] = genX.next()
      yield [Xi,Xi],Yi
```

```python
def training_and_evaluation(model, model_id, scheduler):
    # Compile Model
    model.compile(loss='categorical_crossentropy',
                  optimizer=adam,
                  metrics=['accuracy'])
    
    history = model.fit(#multiple_data_generator(datagen, x_train, y_train, batch_size),
                        datagen.flow(x_train, y_train, batch_size=batch_size),
                        steps_per_epoch=len(x_train)/batch_size,
                        epochs=epochs,
                        shuffle=False, # For 2 inputs False, but usually True
                        verbose=1,
                        validation_data=(x_test, y_test),                   
                        callbacks=[configure_callbacks(model_id), scheduler])
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    print('Test loss:', loss)
    print('Test accuracy:', accuracy)
    print('Test error:', (1-accuracy))
    
    return model
```

```python
def CBGN_FUNC(model, filters, gaussian_noise, kernel=(3,3)):
    x = Conv2D(filters, kernel, padding='same')(model)
    x = BN()(x)
    x = GN(gaussian_noise)(x)
    x = Activation('relu')(x)
    
    return x
```

```python
# Outer product with custom image size and depth
def outer_product(x, i_width, i_height, depth):
    phi_I = tf.einsum('ijkm,ijkn->imn',x[0],x[1]) # Einstein Notation  [batch,31,31,depth] x [batch,31,31,depth] -> [batch,depth,depth]
    phi_I = tf.reshape(phi_I,[-1,depth*depth])        # Reshape from [batch_size,depth,depth] to [batch_size, depth*depth]
    phi_I = tf.divide(phi_I,i_width*i_height)                # Divide by feature map size [sizexsize]

    y_ssqrt = tf.multiply(tf.sign(phi_I),tf.sqrt(tf.abs(phi_I)+1e-12)) # Take signed square root of phi_I
    z_l2 = tf.nn.l2_normalize(y_ssqrt)                                 # Apply l2 normalization
    return z_l2
```

## Models

```python
# VGG mini
inputs = keras.Input(shape=x_train.shape[1:], name="input")

x = CBGN_FUNC(inputs, 32, 0.1)
x = CBGN_FUNC(x, 32, 0.2)

x = MaxPooling2D(pool_size=(2, 2), name='vgg_mini_block1')(x)

x = CBGN_FUNC(x, 64, 0.3)
x = CBGN_FUNC(x, 64, 0.3)

x = MaxPooling2D(pool_size=(2, 2), name='vgg_mini_block2')(x)

x = CBGN_FUNC(x, 128, 0.3)
x = CBGN_FUNC(x, 128, 0.3)

x = MaxPooling2D(pool_size=(2, 2), name='vgg_mini_block3')(x)

x = CBGN_FUNC(x, 256, 0.3)
x = CBGN_FUNC(x, 256, 0.3)
x = CBGN_FUNC(x, 256, 0.3)

x = MaxPooling2D(pool_size=(2, 2), name='vgg_mini_block4')(x)

x = CBGN_FUNC(x, 512, 0.3)
x = CBGN_FUNC(x, 512, 0.3)
x = CBGN_FUNC(x, 256, 0.3)

x = MaxPooling2D(pool_size=(2, 2), name='vgg_mini_block5')(x)

d1 = Dropout(0.5, name='d1')(x)
d2 = Dropout(0.5, name='d2')(x)

width, height, depth = d1.shape[1:]

arguments = {'i_width': width, 'i_height':height, 'depth':depth}

x = Lambda(outer_product, arguments=arguments, name='outer_product')([d1, d2])
x = Dropout(0.5)(x)

outputs = Dense(num_classes, activation='softmax', name='main_output')(x)

model_vgg_mini_fully_bilinear = keras.Model(inputs=inputs, outputs=outputs, name='model_vgg_mini_fully_bilinear')
model_vgg_mini_fully_bilinear.summary()
```

```python
# VGG16 pre-trained
model = VGG16(weights='imagenet',
              include_top=False,
              input_shape=x_train.shape[1:])
model.summary()

for layer in model.layers:
    layer.trainable = False

conv = model.get_layer("block5_pool")

d1 = Dropout(0.5)(conv.output)
d2 = Dropout(0.5)(conv.output)

width, height, depth = d1.shape[1:]

arguments = {'i_width': width, 'i_height':height, 'depth':depth}

x = Lambda(outer_product, arguments=arguments, name='outer_product')([d1, d2])
x = Dropout(0.5)(x)

outputs = Dense(num_classes, activation='softmax', name='out')(x)

model_vgg16_fully_bilinear = keras.Model(inputs=model.input, outputs=outputs) 
model_vgg16_fully_bilinear.summary()
```
